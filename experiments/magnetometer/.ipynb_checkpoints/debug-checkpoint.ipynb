## Model

```python
# Inducing variable
n_inducing_variable = 100
n_grid_inducing_variable = int(np.sqrt(n_inducing_variable).item())
xx, yy = np.linspace(-2, 5, n_grid_inducing_variable), np.linspace(-4, 2, n_grid_inducing_variable)
z1, z2 = np.meshgrid(xx, yy)
zz = np.vstack((z1.flatten(), z2.flatten())).T
inducing_variable = zz.tolist()
```

```python
plt.scatter(zz[:, 0], zz[:, 1])
plt.xlim((-2, 5))
plt.ylim((-4, 2))
plt.show()
```

```python
# Model
likelihood = gpflow.likelihoods.Gaussian(variance=0.1)
kernel = gpflow.kernels.Sum(kernels=[gpflow.kernels.Constant(), gpflow.kernels.Matern52()])
model = t_SVGP_cont(kernel=kernel, likelihood=likelihood, inducing_variable=inducing_variable, num_latent_gps=1)
model.kernel.kernels[0].variance.assign(500)
gpflow.set_trainable(model.inducing_variable.Z, True)
```

```python
optimizer = tf.optimizers.Adam(learning_rate=0.01)
memory = None
online_gp = OnlineGP(model=model, memory=memory, opt_hypers=optimizer, Z_picker=fixed_Z, memory_picker=random_picker, 
                     n_steps=10, lambda_lr=0.1, num_mem=100)
```

## tSVGP model

```python
likelihood = gpflow.likelihoods.Gaussian(variance=0.1)
kernel = gpflow.kernels.Sum(kernels=[gpflow.kernels.Constant(), gpflow.kernels.Matern52()])
tsvgp_model = t_SVGP_white(kernel=kernel, likelihood=likelihood, inducing_variable=inducing_variable, 
                           num_latent_gps=1)
tsvgp_model.kernel.kernels[0].variance.assign(500)
gpflow.set_trainable(tsvgp_model.inducing_variable.Z, True)
optimizer = tf.optimizers.Adam(learning_rate=0.01)
```

```python
data = (streaming_train_data[0][0],streaming_train_data[0][1])
batch_size = data[0].shape[0]
```

```python
_, _, _ = optimize_full_model(tsvgp_model, train_data=data, test_data=test_data, optimizer=optimizer,
                              minibatch_size=batch_size, iterations=4000, lambda_lr=0.8, debug=True)
```

```python
optimized_z_tsvgp = tsvgp_model.inducing_variable.Z.numpy()
```

```python
fig, axs = plt.subplots(1, 1)

plt.scatter(optimized_z_tsvgp[:, 0], optimized_z_tsvgp[:, 1], color="black")
plt.plot(data[0][:, 0], data[0][:, 1])


axs.set_aspect("equal")
plt.xlim((-2, 5))
plt.ylim((-4, 2))
plt.show()
```
